```python
from TTS.config import load_config
from TTS.utils.manage import ModelManager
from TTS.utils.synthesizer import Synthesizer
```

### Prepare config variables

```python
model_path = "best_model.pth"
config_path = "config.json"
speakers_file_path = "speakers.pth"
```

### Synthesizer Instance

```python
synthesizer = Synthesizer(
    model_path, 
    config_path, 
    speakers_file_path, 
    None, None, None,
)
```

### Synthesizing Text to Speech

First define the text you want to synthesize and the speaker

```python
# here write the text you want to transform into speech
text = "Moien, wéi geet et? Wéi ass d'Wieder? Wou liefs du? Wéi al bass du?"

# choose a speaker between Judith, Luc, Sara or Charel
speaker_name = "Judith"

# keep language set to x-lb
language_name = "x-lb"
```

now we convert speech to text

```python
wavs = synthesizer.tts(
    text, 
    speaker_name,
    language_name='x-lb'
)
```
